```python
import matplotlib.pyplot as plt
import pandas as pd

# Carregando os dados
dados_climaticos = pd.read_csv('dados_climaticos.csv')

# Tarefa 1: Análise de Temperatura Média Mensal
dados_climaticos['Data'] = pd.to_datetime(dados_climaticos['Data'])
dados_climaticos['Mês'] = dados_climaticos['Data'].dt.month
temp_media_mensal = dados_climaticos.groupby('Mês')['Temperatura Máxima'].mean()

plt.figure(figsize=(10, 6))
plt.plot(temp_media_mensal.index, temp_media_mensal, marker='o', linestyle='-', color='blue')
plt.title('Temperatura Média Mensal')
plt.xlabel('Mês')
plt.ylabel('Temperatura Média (°C)')
plt.grid(True)
plt.show()

# Tarefa 2: Análise de Precipitação Anual
dados_climaticos['Ano'] = dados_climaticos['Data'].dt.year
precipitacao_anual = dados_climaticos.groupby('Ano')['Precipitação'].sum()

plt.figure(figsize=(10, 6))
plt.bar(precipitacao_anual.index, precipitacao_anual, color='skyblue')
plt.title('Precipitação Anual')
plt.xlabel('Ano')
plt.ylabel('Precipitação Total (mm)')
plt.show()










# Tarefa 3: Relação entre Temperatura Máxima e Velocidade do Vento
plt.figure(figsize=(10, 6))
plt.scatter(dados_climaticos['Temperatura Máxima'], dados_climaticos['Velocidade do Vento'], alpha=0.5)
plt.title('Relação entre Temperatura Máxima e Velocidade do Vento')
plt.xlabel('Temperatura Máxima (°C)')
plt.ylabel('Velocidade do Vento (km/h)')
plt.grid(True)
plt.show()
```
