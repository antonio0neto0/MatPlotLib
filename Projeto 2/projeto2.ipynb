```python
import matplotlib.pyplot as plt
import pandas as pd

# Carregando os dados
dados_climaticos = pd.read_csv('dados_climaticos.csv')

# Tarefa 1: Anlise de Temperatura Mdia Mensal
dados_climaticos['Data'] = pd.to_datetime(dados_climaticos['Data'])
dados_climaticos['Ms'] = dados_climaticos['Data'].dt.month
temp_media_mensal = dados_climaticos.groupby('Ms')['Temperatura Mxima'].mean()

plt.figure(figsize=(10, 6))
plt.plot(temp_media_mensal.index, temp_media_mensal, marker='o', linestyle='-', color='blue')
plt.title('Temperatura Mdia Mensal')
plt.xlabel('Ms')
plt.ylabel('Temperatura Mdia (C)')
plt.grid(True)
plt.show()

# Tarefa 2: Anlise de Precipitao Anual
dados_climaticos['Ano'] = dados_climaticos['Data'].dt.year
precipitacao_anual = dados_climaticos.groupby('Ano')['Precipitao'].sum()

plt.figure(figsize=(10, 6))
plt.bar(precipitacao_anual.index, precipitacao_anual, color='skyblue')
plt.title('Precipitao Anual')
plt.xlabel('Ano')
plt.ylabel('Precipitao Total (mm)')
plt.show()










# Tarefa 3: Relao entre Temperatura Mxima e Velocidade do Vento
plt.figure(figsize=(10, 6))
plt.scatter(dados_climaticos['Temperatura Mxima'], dados_climaticos['Velocidade do Vento'], alpha=0.5)
plt.title('Relao entre Temperatura Mxima e Velocidade do Vento')
plt.xlabel('Temperatura Mxima (C)')
plt.ylabel('Velocidade do Vento (km/h)')
plt.grid(True)
plt.show()
```
