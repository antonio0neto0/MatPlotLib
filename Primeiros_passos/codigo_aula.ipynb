```python
import matplotlib.pyplot as plt
# Exemplo de Dados
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho']
vendas = [150, 200, 170, 180, 160, 175]

# Gráfica de linha
# Eixo X, Eixo Y
plt.figure(figsize=(7,3),frameon=True,facecolor='lightblue',dpi=150) 
plt.plot(meses, vendas,label='vendas',marker='o',linestyle='dashed',color='#04c94c')
plt.title('Vendas Mensais no primeiro Semestre')
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.legend(title='Legenda',loc='lower right') # upper right, upper left, lower right, lower left
plt.grid(True)
plt.show()
```
