```python
import matplotlib.pyplot as plt
import pandas as pd

# Exiba o total de vendas de cada produto cadastrado no arquivo pedidos.csv
pedidos = pd.read_csv('./pedidos.csv')
pedidos['valor_venda'] = pedidos['quantidade'] * pedidos['preco_unitario'] # obtendo o valor total da venda
pedidos.head(5)

# Extrair o total de vendas de bicicletas
total_pedidos_bicicletas = pedidos[pedidos['produto'] == 'Bicicleta']['valor_venda'].sum()
# Extrair o total de vendas de relógios
total_pedidos_relogios = pedidos[pedidos['produto'] == 'Relógio']['valor_venda'].sum()
# Gráfico de barra normal(vertical)
plt.figure(figsize=(10,5)) # primeiro, mostrar que sem o tamanho devido, não fará sentido
plt.bar(['Total Pedidos Bicicleta','Total Pedidos Relógios'],[total_pedidos_bicicletas,total_pedidos_relogios],color=['blue','green'])
plt.xlabel('Produtos')
plt.ylabel('Preços')
plt.show()
```
