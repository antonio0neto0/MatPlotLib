```python

import matplotlib.pyplot as plt
import pandas as pd

pedidos = pd.read_csv('./pedidos.csv')

# Clculo do valor da venda
pedidos['valor_venda'] = pedidos['quantidade'] * pedidos['preco_unitario']
# Agrupamento por categoria e clculo do total da venda por categoria,alm da ordenao dos resultados com base no valor da venda
vendas_por_categoria = pedidos.groupby(['categoria']).agg({'valor_venda':'sum'}).sort_values(by='valor_venda').reset_index()

# Grfico de barra vertical
plt.figure(figsize=(6,5),dpi=150)
plt.bar(vendas_por_categoria['categoria'],vendas_por_categoria['valor_venda'])
plt.xlabel('Produto')
plt.ylabel('Total de Vendas')
plt.show()

# Grfico de barra horizontal
plt.figure(figsize=(10,5),dpi=150)
plt.barh(vendas_por_categoria['categoria'],vendas_por_categoria['valor_venda'])
plt.ylabel('Produto')
plt.xlabel('Total de Vendas')
plt.show()
```
