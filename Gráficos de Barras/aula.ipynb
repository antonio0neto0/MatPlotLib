Tipos de Gráficos: Gráficos de Barras(Bar Chart)

```python
import matplotlib.pyplot as plt

# Exemplo Simples
produtos = ['Mochila', 'Fones de Ouvido', 'Relógio']
precos = [150, 200, 50]

plt.figure(figsize=(5,5),dpi=150)
# plt.bar(produtos,precos,width=0.4,color='blue')
plt.bar(produtos,precos,width=0.4,color=['red','blue','green'])
plt.xlabel('Produtos')
plt.ylabel('Preços')
plt.show()
```

```python
import matplotlib.pyplot as plt

produtos = ['Mochila', 'Fones de Ouvido', 'Relógio']
precos_2020 = [150, 200, 50]
precos_2030 = [250, 350, 80]

plt.figure(figsize=(5,5),dpi=150)
plt.bar(produtos,precos_2020,color='blue',width=0.4,label='2020')
plt.bar(produtos,precos_2030,color='red',width=0.2,align='edge',label='2030')
plt.xlabel('Produtos')
plt.ylabel('Preços')
plt.legend(title='Anos')
plt.show()
```

```python
# Exiba o total de vendas de cada produto cadastrado no arquivo pedidos.csv
import matplotlib.pyplot as plt
import pandas as pd

pedidos = pd.read_csv('./pedidos.csv')
# Obtendo o total da venda
pedidos['valor_venda'] = pedidos['quantidade'] * pedidos['preco_unitario']
# Agrupar por produto e valor de venda
total_de_vendas_por_produto = pedidos.groupby('produto').agg({'valor_venda':'sum'}).sort_values(by='valor_venda').reset_index()
# Gráfico de barra padrão(vertical)
plt.figure(figsize=(35,5),dpi=150)
plt.bar(total_de_vendas_por_produto['produto'],total_de_vendas_por_produto['valor_venda'])
plt.xlabel('Produtos')
plt.ylabel('Preços')
plt.show()
# Gráfica de barra horizontal
plt.barh(total_de_vendas_por_produto['produto'],total_de_vendas_por_produto['valor_venda'])
plt.xlabel('Produtos')
plt.ylabel('Preços')
plt.show()
```
