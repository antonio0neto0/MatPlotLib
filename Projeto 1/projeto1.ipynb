```python
import matplotlib.pyplot as plt
import pandas as pd

# Tarefa 1: Análise de Performance por Categoria
vendas = pd.read_csv('dados_vendas.csv')
# Agrupando os dados por 'Categoria' e calculando o total de vendas
vendas_por_categoria = vendas.groupby('Categoria')['Vendas'].sum()

# Preparando dados para o gráfico de pizza
total_vendas = vendas_por_categoria.values
categorias = vendas_por_categoria.index

# Criando o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(total_vendas, labels=categorias, autopct='%1.1f%%', startangle=90, colors=['lightblue','green','yellow','red'])
plt.title('Participação Percentual de Vendas por Categoria')
plt.show()


# Tarefa 2: Tendências de Vendas Mensais
# Extraindo o mês da data do pedido

# Certifique-se de que 'Data do Pedido' é um datetime
vendas['Data do Pedido'] = pd.to_datetime(vendas['Data do Pedido'])
vendas['Mês'] = vendas['Data do Pedido'].dt.month
# Calculando o total de vendas por mês
vendas_mensais = vendas.groupby('Mês')['Vendas'].sum()
meses = vendas_mensais.index
totais_mensais = vendas_mensais.values
# Criando o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(meses, totais_mensais, marker='o', linestyle='-', color='darkblue')
plt.title('Tendência de Vendas Mensais')
plt.xlabel('Mês')
plt.ylabel('Total de Vendas')
plt.grid(True)
plt.show()


# Tarefa 3: Impacto dos Finais de Semana nas Vendas
# Identificando dias úteis (0-4) e finais de semana (5-6)
vendas['Dia da Semana'] = vendas['Data do Pedido'].dt.dayofweek
vendas_dias_uteis = vendas[vendas['Dia da Semana'] < 5]['Vendas'].sum()
vendas_finais_semana = vendas[vendas['Dia da Semana'] >= 5]['Vendas'].sum()
# Criando gráficos de barras para comparar vendas em dias úteis e finais de semana
plt.figure(figsize=(10, 6))
plt.bar(['Dias Úteis', 'Finais de Semana'], [vendas_dias_uteis, vendas_finais_semana], color=['green', 'red'])
plt.title('Comparação de Vendas: Dias Úteis vs. Finais de Semana')
plt.ylabel('Total de Vendas')
plt.show()

```
