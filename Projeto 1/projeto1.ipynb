```python
import matplotlib.pyplot as plt
import pandas as pd

# Tarefa 1: Anlise de Performance por Categoria
vendas = pd.read_csv('dados_vendas.csv')
# Agrupando os dados por 'Categoria' e calculando o total de vendas
vendas_por_categoria = vendas.groupby('Categoria')['Vendas'].sum()

# Preparando dados para o grfico de pizza
total_vendas = vendas_por_categoria.values
categorias = vendas_por_categoria.index

# Criando o grfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(total_vendas, labels=categorias, autopct='%1.1f%%', startangle=90, colors=['lightblue','green','yellow','red'])
plt.title('Participao Percentual de Vendas por Categoria')
plt.show()


# Tarefa 2: Tendncias de Vendas Mensais
# Extraindo o ms da data do pedido

# Certifique-se de que 'Data do Pedido'  um datetime
vendas['Data do Pedido'] = pd.to_datetime(vendas['Data do Pedido'])
vendas['Ms'] = vendas['Data do Pedido'].dt.month
# Calculando o total de vendas por ms
vendas_mensais = vendas.groupby('Ms')['Vendas'].sum()
meses = vendas_mensais.index
totais_mensais = vendas_mensais.values
# Criando o grfico de linha
plt.figure(figsize=(10, 6))
plt.plot(meses, totais_mensais, marker='o', linestyle='-', color='darkblue')
plt.title('Tendncia de Vendas Mensais')
plt.xlabel('Ms')
plt.ylabel('Total de Vendas')
plt.grid(True)
plt.show()


# Tarefa 3: Impacto dos Finais de Semana nas Vendas
# Identificando dias teis (0-4) e finais de semana (5-6)
vendas['Dia da Semana'] = vendas['Data do Pedido'].dt.dayofweek
vendas_dias_uteis = vendas[vendas['Dia da Semana'] < 5]['Vendas'].sum()
vendas_finais_semana = vendas[vendas['Dia da Semana'] >= 5]['Vendas'].sum()
# Criando grficos de barras para comparar vendas em dias teis e finais de semana
plt.figure(figsize=(10, 6))
plt.bar(['Dias teis', 'Finais de Semana'], [vendas_dias_uteis, vendas_finais_semana], color=['green', 'red'])
plt.title('Comparao de Vendas: Dias teis vs. Finais de Semana')
plt.ylabel('Total de Vendas')
plt.show()

```
