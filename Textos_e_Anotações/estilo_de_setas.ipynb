```python
import matplotlib.pyplot as plt

# Define some data points
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create subplots for each arrow style
fig, axes = plt.subplots(3, 3, figsize=(12, 12))

# Arrow Style: '->'
ax1 = axes[0, 0]
ax1.plot(x, y, 'bo')
ax1.annotate('Arrow Style: ->', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='->'))
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Arrow Style: ->')

# Arrow Style: '-['
ax2 = axes[0, 1]
ax2.plot(x, y, 'bo')
ax2.annotate('Arrow Style: -[', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='-['))
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Arrow Style: -[')

# Arrow Style: '|-|'
ax3 = axes[0, 2]
ax3.plot(x, y, 'bo')
ax3.annotate('Arrow Style: |-|', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='|-|'))
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_title('Arrow Style: |-|')

# Arrow Style: '-|>'
ax4 = axes[1, 0]
ax4.plot(x, y, 'bo')
ax4.annotate('Arrow Style: -|>', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='-|>'))
ax4.set_xlabel('X')
ax4.set_ylabel('Y')
ax4.set_title('Arrow Style: -|>')

# Arrow Style: '<-'
ax5 = axes[1, 1]
ax5.plot(x, y, 'bo')
ax5.annotate('Arrow Style: <-', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='<-'))
ax5.set_xlabel('X')
ax5.set_ylabel('Y')
ax5.set_title('Arrow Style: <-')

# Arrow Style: '<->'
ax6 = axes[1, 2]
ax6.plot(x, y, 'bo')
ax6.annotate('Arrow Style: <->', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='<->'))
ax6.set_xlabel('X')
ax6.set_ylabel('Y')
ax6.set_title('Arrow Style: <->')

# Arrow Style: 'fancy'
ax7 = axes[2, 0]
ax7.plot(x, y, 'bo')
ax7.annotate('Arrow Style: fancy', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='fancy'))
ax7.set_xlabel('X')
ax7.set_ylabel('Y')
ax7.set_title('Arrow Style: fancy')

# Arrow Style: 'simple'
ax8 = axes[2, 1]
ax8.plot(x, y, 'bo')
ax8.annotate('Arrow Style: simple', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='simple'))
ax8.set_xlabel('X')
ax8.set_ylabel('Y')
ax8.set_title('Arrow Style: simple')

# Arrow Style: 'wedge'
ax9 = axes[2, 2]
ax9.plot(x, y, 'bo')
ax9.annotate('Arrow Style: wedge', xy=(3, 9), xytext=(3.5, 12),
             arrowprops=dict(arrowstyle='wedge'))
ax9.set_xlabel('X')
ax9.set_ylabel('Y')
ax9.set_title('Arrow Style: wedge')

plt.tight_layout()
plt.show()
```
