Tipos de Gráficos: Gráficos de Pizza(Pie Chart)

```python
import matplotlib.pyplot as plt

# Dados de exemplo
categorias = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
valores = [10, 20, 30, 40]

plt.figure(figsize=(7,7),dpi=150)
# 'autopct' permite mostrar a porcentagem de cada fatia
# Você pode alterar o autopct da seguinte forma: 
# %1.1f%% para 1 casa decimal após o ponto (ex: 15.5%)
# %1.2f%% para 2 casas decimais após o ponto (ex: 15.55%)
# %1.3f%% para 3 casas decimais após o ponto (ex: 15.555%)
cores = ['gold','lightblue','blue','lightgreen','purple']
explode = (0.1,0,0,0)
plt.pie(x=valores,labels=categorias,autopct='%1.2f%%',startangle=90,shadow=True,explode=explode,colors=cores,textprops={'color':'white','backgroundcolor':'black','weight':'demi'}) # regular, demi, bold
plt.title('Distribuição por categorias')
plt.show()
```
