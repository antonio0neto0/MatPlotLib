```python
import matplotlib.pyplot as plt
import pandas as pd
# Ler arquivo e mostrar quais foram os produtos mais vendidos

pedidos = pd.read_csv('vendas_instagram.csv')
# Como irei agrupar as informações utilizando pandas
total_pedidos = pedidos.groupby('produto')['preco'].sum()
total_pedidos
# Como exibir isso, usando matplotlib
plt.figure(figsize=(15,5),facecolor='lightblue',frameon=True,dpi=150)
plt.plot(total_pedidos.index,total_pedidos.values)
plt.title('Vendas mensais no 1° Semestre')
plt.xlabel('Produtos')
plt.ylabel('Preços')
plt.grid(True)
plt.show()
```
