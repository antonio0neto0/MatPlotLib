```python
import matplotlib.pyplot as plt
import numpy as np

figure, axes = plt.subplots(nrows=2,ncols=2, figsize=(10, 10))

# Dados para o grfico de barra
x = ['A', 'B', 'C', 'D', 'E']
y = [10, 24, 36, 40, 15]

# Cria o grfico de barra
axes[0, 0].bar(x, y)
axes[0, 0].set_title('Grfico de Barra')

# Dados para o grfico de linha
x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)

# Cria o grfico de linha
axes[0, 1].plot(x_line, y_line)
axes[0, 1].set_title('Grfico de Linha')

# Dados para o grfico de disperso
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

# Cria o grfico de disperso
axes[1, 0].scatter(x_scatter, y_scatter)
axes[1, 0].set_title('Grfico de Disperso')

# Dados para o grfico de pizza
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 45, 10, 5]

# Cria o grfico de pizza
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
axes[1, 1].set_title('Grfico de Pizza')

plt.show()
```
