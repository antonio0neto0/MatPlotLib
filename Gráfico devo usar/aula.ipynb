# Qual Tipo de Gráfico devo usar?

# Gráficos de Linha (Line Chart)
- São ideais para visualizar tendências ao longo do tempo ou a relação entre duas variáveis. Eles são úteis para dados temporais ou sequenciais.

```python
import matplotlib.pyplot as plt

# Dados de exemplo
meses = ['jan','fev','mar','abr','mai']
vendas = [250,530,234,456,745]

# Plotagem do gráfico de linha
plt.plot(meses, vendas)
plt.title('Gráfico de Linha')
plt.xlabel('Vendas')
plt.ylabel('Meses')
plt.grid(True)
plt.show()
```

# Gráficos de Barras (Bar Chart)
- Usados para comparar quantidades entre diferentes categorias. Eles são ótimos para visualizar diferenças em quantidades ou frequências entre grupos.

```python
import matplotlib.pyplot as plt

# Dados de exemplo
cidades = ['São Paulo', 'Rio de Janeiro', 'Brasília', 'Salvador', 'Fortaleza']
populacao = [12180000, 6710000, 3050000, 2886000, 2800000]

# Plotagem do gráfico de barras
plt.figure(figsize=(7,3))
plt.bar(cidades, populacao)
plt.title('População das Maiores Cidades do Brasil')
plt.xlabel('Cidade')
plt.ylabel('População')
plt.show()
```

# Histogramas (Histograms)
- Utilizados para mostrar a distribuição de uma variável contínua, permitindo identificar padrões como normalidade, assimetria, ou presença de outliers.

```python
import matplotlib.pyplot as plt

# Dados de exemplo (idade de uma pequena população)
idade_populacao = [22, 25, 30, 35, 40, 42, 45, 50, 55, 60, 65, 70, 75, 80, 85]

# Plotagem do histograma
plt.hist(idade_populacao, bins=5, edgecolor='black')
plt.title('Distribuição da Idade da População')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

# Gráficos de Dispersão (Scatter Plots)
- Ideais para explorar a relação entre duas variáveis numéricas, podendo indicar a correlação entre elas.

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados de exemplo (altura vs. peso)
altura = np.random.normal(170, 10, 100)
peso = altura * 0.7 + np.random.normal(0, 5, 100)

# Plotagem do gráfico de dispersão
plt.scatter(altura, peso, color='blue')
plt.title('Altura vs. Peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.show()
```

# Gráficos de Pizza (Pie Charts)
- Aplicáveis para mostrar proporções ou porcentagens de um todo, úteis para destacar a contribuição de cada categoria ao total.

```python
import matplotlib.pyplot as plt

# Dados de exemplo
cursos = ['Engenharia', 'Ciências Sociais', 'Medicina', 'Economia']
quantidades_alunos = [300, 150, 450, 200]

# Plotagem do gráfico de pizza
plt.pie(quantidades_alunos, labels=cursos, autopct='%1.1f%%')
plt.title('Distribuição de Alunos por Curso')
plt.show()
```
